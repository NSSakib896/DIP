```python
import cv2 as cv 
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread("hist.png")
plt.imshow(img)
plt.axis("off")
```

```python
lab_image = cv.cvtColor(img, cv.COLOR_BGR2LAB)
plt.imshow(lab_image)
```

```python
l, a, b = cv.split(lab_image)
plt.imshow(l)
```

```python
n,bins,patches = plt.hist(l.flat, bins=255, range=(0, 255))
```

```python
equ = cv.equalizeHist(l)
plt.imshow(equ, cmap="gray")
```

```python
plt.hist(equ.flat, bins=255, range=(0, 255))
```

```python
merged_img = cv.merge((equ, a, b))
equalized_image = cv.cvtColor(merged_img, cv.COLOR_LAB2BGR)
```

```python
plt.imshow(equalized_image)
```

```python
clahe_instance = cv.createCLAHE(clipLimit=5.0)
clahe_l = clahe_instance.apply(l)
clahe_img = cv.merge((clahe_l,a,b))
clahe_img = cv.cvtColor(clahe_img, cv.COLOR_LAB2BGR)
```

```python
plt.imshow(clahe_img)
```
