```python
import numpy as np 
import matplotlib.pyplot as plt
import cv2 as cv 
```

```python
#image loading 
image = cv.imread("cat2.jpg",0)
plt.imshow(image,cmap='gray')
plt.show
```

```python
def formula(x,max):
    c = 255/np.log(1+255)
    return c*np.log(1+x)
```

```python
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        image[i][j] = formula(image[i][j])
```

```python
plt.imshow(image,cmap="gray")
```
