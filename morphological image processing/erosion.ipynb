```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2 as cv
```

```python
image = cv.imread("ErosionDilation.jpg",0)
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        if image[i][j]>200:
            image[i][j]=0
        else:
            image[i][j]=1
    
plt.imshow(image,cmap='gray')
plt.title("input")
```

```python
kernel = np.ones((5,5),np.uint8)
```

```python
erode = cv.erode(image,kernel=kernel,iterations=1)
plt.figure(figsize=(20,40))
plt.subplot(1,2,2)
plt.imshow(erode,cmap="gray")
plt.subplot(1,2,1)
plt.imshow(image,cmap="gray")
plt.title("input")
plt.show()
```

```python
dilate = cv.dilate(image,kernel=kernel,iterations=1)
plt.figure(figsize=(20,40))
plt.subplot(1,2,2)
plt.imshow(dilate,cmap="gray")
plt.subplot(1,2,1)
plt.imshow(image,cmap="gray")
plt.title("input")
plt.show()
```

```python
boundary_extraction = image-erode
plt.figure(figsize=(20,40))
plt.subplot(1,2,2)
plt.imshow(boundary_extraction,cmap="gray")
plt.subplot(1,2,1)
plt.imshow(image,cmap="gray")
plt.title("input")
plt.show()
```

```python
open = cv.dilate(erode,kernel=kernel,iterations=1)
close = cv.erode(dilate,kernel=kernel,iterations=1)
plt.figure(figsize=(20,40))
plt.subplot(1,3,2)
plt.imshow(open,cmap="gray")
plt.axis(False)
plt.subplot(1,3,1)
plt.imshow(image,cmap="gray")
plt.title("input")
plt.axis(False)
plt.subplot(1,3,3)
plt.imshow(close,cmap="gray")
plt.title("closing")
plt.axis(False)
plt.show()
```
