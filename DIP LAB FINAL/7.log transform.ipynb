```python
import numpy as np 
import matplotlib.pyplot as plt
import cv2 as cv 
#image loading 
plt.figure(figsize=(20,20))
plt.subplot(2,1,1)
plt.title('input')
image = cv.imread("cat1.jpg",0)
plt.imshow(image,cmap='gray')

def formula(x):
    c = 255/np.log(1+255)
    return c*np.log(1+x)
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        image[i][j] = formula(image[i][j])
plt.subplot(2,1,2)
plt.title('output')
plt.imshow(image,cmap="gray")
plt.axis(False)
plt.show()
```
