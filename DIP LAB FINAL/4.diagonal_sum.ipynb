```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os 
img= cv.imread("cat1.jpg") 
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.subplot(1,2,1)
plt.title("input")
plt.imshow(img, cmap='gray')
plt.axis('off')
def get_diagonal_sum(img):
    diagonal_1 = np.trace(img)
    diagonal_2 = np.trace(np.fliplr(img))
    diagonal_sum = diagonal_1 + diagonal_2 - img[img.shape[0]//2, img.shape[1]//2]
    return diagonal_sum
get_diagonal_sum(img)
img[img.shape[0]//2, img.shape[1]//2] = get_diagonal_sum(img)
plt.subplot(1,2,2)
plt.title("output")
plt.imshow(img,cmap='gray')
plt.axis('off')
plt.show()
```
